```
df = pd.read_csv('/content/drive/MyDrive/petrol_consumption.csv') # reading the csv file of petrol consumption from google drive
```

```
df # printing the data
```

```
df_x = df.drop('Petrol_Consumption', axis=1) # axis=1 for column wise and axis=0 for row wise
df_x
df_y = df['Petrol_Consumption']
df_y
```

```
df_x
```

```
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.25)
```

```
x_train
```

```
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```
y_predict = model.predict(x_test)
```

```
y_predict
```

```
y_test
```

```
df_prediction = pd.DataFrame({'predicted_petrol_consumption' :y_predict, 'Actual_petrol_consumption' :y_test})
df_prediction
```

```
from sklearn import metrics
metrics.mean_squared_error(y_predict, y_test)
```
